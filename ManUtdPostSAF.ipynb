```python
import pandas as pd
import numpy as np
data = pd.read_csv('stats.csv')
data.head()
```

```python
#Extracting only Manchester United data.
manUtd_data = data[data.team=='Manchester United']
```

```python
#Extracting out total number of goals scored every season.
goals_scored = ['goals','season'] 
goals_scoredVseason = manUtd_data[goals_scored]
```

```python
#Extracting out total number of goals conceded every season.
goals_conceded = ['goals_conceded','season']
goals_concededVseason = manUtd_data[goals_conceded]
```

```python
import matplotlib.pyplot as plt

#Plotting Wins and Goals Scored v/s Goals Conceded
plt.rcParams['figure.figsize'] = [12, 8] #creates a graph of 10x7 inches
plt.plot(goals_scoredVseason.season, goals_scoredVseason.goals, label = 'Goals Scored', marker='o') #plot Goals Scored vs Season
plt.plot(goals_concededVseason.season, goals_concededVseason.goals_conceded, label = 'Goals Conceded', marker='o') #plot Goals Conceded vs Season, with markers and label
plt.ylim(0,100) #for setting y limits from 15 to 30.
plt.grid(which='major', axis='both', linestyle='-.', linewidth=0.75) #plotting grid. See pyplot.grid in Google.
plt.xticks(rotation=60, fontsize=14) #rotate x axis labels and increase font.
plt.yticks(fontsize=14) #increase yticks fontsize.
plt.legend(fontsize=14) #show labels as legend
plt.title('Goals Scored v/s Goals Conceded', fontsize = 20)

try: #try catch is used because here x-coord is a string like "2006-2007" and hence it throws a TypeError. So TypeError errors are overlooked in except statement.
    
    plt.annotate('Moyes appointed', xy=('2013-2014',64), xytext=('2013-2014', 75),arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8), fontsize =10) 
    #For the goals scored when Moyes was appointed.
    
    plt.annotate('Moyes appointed', xy=('2013-2014',43), xytext=('2013-2014', 54),arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8), fontsize =10) 
    #For the goals conceded when Moyes was appointed.
    
    plt.annotate('Van Gaal appointed', xy=('2014-2015',62), xytext=('2014-2015', 70),arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8), fontsize =10) 
    #For goals scored when LVG was appointed.
    
    plt.annotate('Van Gaal appointed', xy=('2014-2015',37), xytext=('2014-2015', 45),arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8), fontsize =10)
    #For goals conceded when LVG was appointed.
    
    plt.annotate('Mourinho appointed', xy=('2016-2017',54), xytext=('2016-2017', 62),arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8), fontsize =10) 
    #For goals scored when Mou was appointed. 
    
    plt.annotate('Mourinho appointed', xy=('2016-2017',30), xytext=('2016-2017', 38),arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8), fontsize =10) 
    #For goals conceded when Mou was appointed. 
    
    plt.text('2009-2010',50,'The SAF era', horizontalalignment='center', verticalalignment='center', fontsize =16, bbox=dict(facecolor='red', alpha=0.5)) 
    #For the middle big text
except TypeError: 
    pass
```
